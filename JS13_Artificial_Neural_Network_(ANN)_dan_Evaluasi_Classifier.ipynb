# Praktikum 1

## Langkah:

1. Buat dataset sederhana (XOR).
2. Inisialisasi bobot dan bias.
3. Implementasikan forward pass.
4. Hitung error dan lakukan backpropagation.
5. Update bobot menggunakan gradient descent.

```python
import numpy as np

# Dataset XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

# Parameter
input_size = 2
hidden_size = 2
output_size = 1
lr = 0.1

# Inisialisasi bobot
W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))
W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))

# Fungsi aktivasi
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Training
for epoch in range(10000):
    # Forward pass
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)

    # Hitung error
    error = y - a2

    # Backpropagation
    d_a2 = error * sigmoid_derivative(a2)
    d_W2 = np.dot(a1.T, d_a2)
    d_b2 = np.sum(d_a2, axis=0, keepdims=True)

    d_a1 = np.dot(d_a2, W2.T) * sigmoid_derivative(a1)
    d_W1 = np.dot(X.T, d_a1)
    d_b1 = np.sum(d_a1, axis=0, keepdims=True)

    # Update bobot
    W1 += lr * d_W1
    b1 += lr * d_b1
    W2 += lr * d_W2
    b2 += lr * d_b2

    if epoch % 1000 == 0:
        loss = np.mean(np.square(error))
        print(f"Epoch {epoch}, Loss: {loss}")

# Output akhir
print("Prediksi:")
print(a2)
```

```python
import numpy as np

# Dataset XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

# Parameter - UBAH hidden_size menjadi 3
input_size = 2
hidden_size = 3  # Diubah dari 2 menjadi 3
output_size = 1
lr = 0.1

# Inisialisasi bobot
np.random.seed(42)  # Untuk reproduksibilitas
W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))
W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))

# Fungsi aktivasi
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Training
losses_3_neurons = []
for epoch in range(10000):
    # Forward pass
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)

    # Hitung error
    error = y - a2

    # Backpropagation
    d_a2 = error * sigmoid_derivative(a2)
    d_W2 = np.dot(a1.T, d_a2)
    d_b2 = np.sum(d_a2, axis=0, keepdims=True)

    d_a1 = np.dot(d_a2, W2.T) * sigmoid_derivative(a1)
    d_W1 = np.dot(X.T, d_a1)
    d_b1 = np.sum(d_a1, axis=0, keepdims=True)

    # Update bobot
    W1 += lr * d_W1
    b1 += lr * d_b1
    W2 += lr * d_W2
    b2 += lr * d_b2

    loss = np.mean(np.square(error))
    losses_3_neurons.append(loss)

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

# Output akhir
print("\nPrediksi dengan 3 neuron hidden layer:")
print(a2)
print(f"\nFinal Loss: {losses_3_neurons[-1]}")
```

### 3. Implementasi dengan Fungsi Aktivasi ReLU

```python
# Fungsi aktivasi ReLU
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)

# Training dengan ReLU (hidden layer 3 neurons)
np.random.seed(42)
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

input_size = 2
hidden_size = 3
output_size = 1
lr = 0.1

W1_relu = np.random.randn(input_size, hidden_size)
b1_relu = np.zeros((1, hidden_size))
W2_relu = np.random.randn(hidden_size, output_size)
b2_relu = np.zeros((1, output_size))

losses_relu = []
for epoch in range(10000):
    # Forward pass - ReLU di hidden layer, Sigmoid di output
    z1 = np.dot(X, W1_relu) + b1_relu
    a1 = relu(z1)  # ReLU di hidden layer
    z2 = np.dot(a1, W2_relu) + b2_relu
    a2 = sigmoid(z2)  # Sigmoid di output layer

    # Hitung error
    error = y - a2

    # Backpropagation
    d_a2 = error * sigmoid_derivative(a2)
    d_W2 = np.dot(a1.T, d_a2)
    d_b2 = np.sum(d_a2, axis=0, keepdims=True)

    d_a1 = np.dot(d_a2, W2_relu.T) * relu_derivative(z1)  # Gunakan z1 untuk ReLU derivative
    d_W1 = np.dot(X.T, d_a1)
    d_b1 = np.sum(d_a1, axis=0, keepdims=True)

    # Update bobot
    W1_relu += lr * d_W1
    b1_relu += lr * d_b1
    W2_relu += lr * d_W2
    b2_relu += lr * d_b2

    loss = np.mean(np.square(error))
    losses_relu.append(loss)

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

print("\nPrediksi dengan ReLU (3 neurons):")
print(a2)
print(f"\nFinal Loss: {losses_relu[-1]}")
```

```python
# MLP regresi (Keras)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# 1. Load (menggunakan California Housing dataset sebagai pengganti Boston)
data = fetch_california_housing()
X = data.data; y = data.target

# 2. Preprocess
scaler = StandardScaler(); Xs = scaler.fit_transform(X)
X_train, X_val, y_train, y_val = train_test_split(Xs, y, test_size=0.2, random_state=42)

# 3. Build model
model = Sequential([
Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
Dense(32, activation='relu'),
Dense(1)
])
model.compile(optimizer=Adam(1e-3), loss='mse', metrics=['mae'])

# 4. Train
h = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=200, batch_size=32, verbose=0)

# 5. Plot
plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.plot(h.history['loss'], label='train_loss'); plt.plot(h.history['val_loss'], label='val_loss'); plt.legend(); plt.title('MSE')
plt.subplot(1,2,2); plt.plot(h.history['mae'], label='train_mae'); plt.plot(h.history['val_mae'], label='val_mae'); plt.legend(); plt.title('MAE')
plt.show()

from sklearn.metrics import mean_squared_error
pred = model.predict(X_val)
print('RMSE:', np.sqrt(mean_squared_error(y_val, pred)))
```

## Konfigurasi Awal (128-64 neurons dengan ReLU)

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import numpy as np
import time

# 1. Load dataset MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 → 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

print(f"Shape data training: {X_train.shape}")
print(f"Shape label training: {y_train.shape}")
print(f"Shape data testing: {X_test.shape}")
print(f"Shape label testing: {y_test.shape}")

# 2. Bangun model JST (Konfigurasi Awal: 128-64 neurons)
np.random.seed(42)
tf.random.set_seed(42)

model_base = Sequential([
    Flatten(input_shape=(28, 28)),  # Ubah gambar 28x28 menjadi vektor 784
    Dense(128, activation='relu'),   # Hidden layer 1
    Dense(64, activation='relu'),    # Hidden layer 2
    Dense(10, activation='softmax')  # Output layer (10 kelas)
])

# 3. Kompilasi model
model_base.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\nModel Summary:")
model_base.summary()

# 4. Latih model
print("\nTraining model dengan konfigurasi awal (128-64 neurons, ReLU)...")
start_time = time.time()
history_base = model_base.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=10,
    batch_size=128,
    verbose=1
)
training_time_base = time.time() - start_time

# 5. Evaluasi model
loss_base, acc_base = model_base.evaluate(X_test, y_test, verbose=0)
print(f"\nAkurasi pada data uji: {acc_base:.4f}")
print(f"Loss pada data uji: {loss_base:.4f}")
print(f"Waktu pelatihan: {training_time_base:.2f} detik")
```

## Eksperimen 1: Ubah Jumlah Neuron (256-128 neurons)

```python
# Model dengan 256-128 neurons
np.random.seed(42)
tf.random.set_seed(42)

model_256_128 = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(256, activation='relu'),   # Hidden layer 1 - lebih banyak neuron
    Dense(128, activation='relu'),   # Hidden layer 2 - lebih banyak neuron
    Dense(10, activation='softmax')
])

model_256_128.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("Model Summary (256-128 neurons):")
model_256_128.summary()

print("\nTraining model dengan 256-128 neurons...")
start_time = time.time()
history_256_128 = model_256_128.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=10,
    batch_size=128,
    verbose=1
)
training_time_256_128 = time.time() - start_time

loss_256_128, acc_256_128 = model_256_128.evaluate(X_test, y_test, verbose=0)
print(f"\nAkurasi: {acc_256_128:.4f}")
print(f"Loss: {loss_256_128:.4f}")
print(f"Waktu pelatihan: {training_time_256_128:.2f} detik")
```

## Eksperimen 2: Tambah Hidden Layer (3 hidden layers)

```python
# Model dengan 3 hidden layers (128-64-32 neurons)
np.random.seed(42)
tf.random.set_seed(42)

model_3_layers = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),   # Hidden layer 1
    Dense(64, activation='relu'),    # Hidden layer 2
    Dense(32, activation='relu'),    # Hidden layer 3 - layer tambahan
    Dense(10, activation='softmax')
])

model_3_layers.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("Model Summary (3 hidden layers: 128-64-32):")
model_3_layers.summary()

print("\nTraining model dengan 3 hidden layers...")
start_time = time.time()
history_3_layers = model_3_layers.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=10,
    batch_size=128,
    verbose=1
)
training_time_3_layers = time.time() - start_time

loss_3_layers, acc_3_layers = model_3_layers.evaluate(X_test, y_test, verbose=0)
print(f"\nAkurasi: {acc_3_layers:.4f}")
print(f"Loss: {loss_3_layers:.4f}")
print(f"Waktu pelatihan: {training_time_3_layers:.2f} detik")
```

## Eksperimen 3: Sigmoid vs ReLU

```python
# Model dengan Sigmoid activation
np.random.seed(42)
tf.random.set_seed(42)

model_sigmoid = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='sigmoid'),  # Sigmoid activation
    Dense(64, activation='sigmoid'),   # Sigmoid activation
    Dense(10, activation='softmax')
])

model_sigmoid.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("Training model dengan Sigmoid activation...")
start_time = time.time()
history_sigmoid = model_sigmoid.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=10,
    batch_size=128,
    verbose=1
)
training_time_sigmoid = time.time() - start_time

loss_sigmoid, acc_sigmoid = model_sigmoid.evaluate(X_test, y_test, verbose=0)
print(f"\nAkurasi (Sigmoid): {acc_sigmoid:.4f}")
print(f"Loss (Sigmoid): {loss_sigmoid:.4f}")
print(f"Waktu pelatihan: {training_time_sigmoid:.2f} detik")
```
